```python
import pandas as pd
import numpy as np
import pprint as pp 
import matplotlib.pyplot as plt
from scipy.stats import mode
from collections import Counter
from operator import itemgetter
import seaborn as sns

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import label_binarize
from sklearn.metrics import accuracy_score, log_loss, confusion_matrix, roc_auc_score, \
                            roc_curve, precision_recall_curve, confusion_matrix, \
                            precision_score, recall_score, f1_score, classification_report
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, RandomForestClassifier, GradientBoostingClassifier, VotingClassifier
from sklearn.cross_validation import cross_val_score

%matplotlib inline
```

```python
by_state_crime = pd.read_csv('by_state_crime_sql.csv')
by_state_funding = pd.read_csv('by_state_funding_information_sql.csv')
by_zipcode_info = pd.read_csv('by_zipcode_information_sql.csv')
hospital_general = pd.read_csv('Hospital_general_sql.csv')
hospital_perform = pd.read_csv('Hospital_performance_sql.csv')
geocode = pd.read_csv('addr_r_coord.csv')
by_state_cancer = pd.read_csv('by_state_cancer_rate.csv')
```

```python
## Joining the two hospital data tables by the 'id' number
##########################################################
hospital_general['state2'] = hospital_general['State']

temp = hospital_general.set_index('id').join(hospital_perform.set_index('id'), how='left', lsuffix='_general', rsuffix='_perform')
temp = temp.dropna().reset_index()
temp2 = temp.set_index('zip_general').join(by_zipcode_info.set_index('Zip Code ZCTA'), how='left')
hospital_joined = temp2.dropna().reset_index()
hospital_joined.rename(columns={'index': 'zip'}, inplace=True)

temp3 = by_state_funding.set_index('state').join(by_state_crime.set_index('state'), how='left', lsuffix='_funding', rsuffix='_crime')
temp4 = temp3.dropna().reset_index().groupby('state_code').agg('mean').reset_index()
temp4.rename(columns={'state_code': 'state'}, inplace=True)
hospital_joined.rename(columns={'State_general': 'state'}, inplace=True)

hospital_joined = hospital_joined.set_index('state2').join(temp4.set_index('state'), how='left')
hospital_joined = hospital_joined.set_index('State_perform').join(by_state_cancer.set_index('State'), how='left')
hospital_joined = hospital_joined.set_index('Address').join(geocode.set_index('Address'), how='left')
hospital_joined = hospital_joined.dropna().reset_index()
hospital_joined.rename(columns={'Rate':'Cancer Rate'}, inplace=True)
hospital_joined.drop(['name_perform', 'address', 'city', 'zip_perform', 'countyName_perform', 'Range'], 1, inplace=True)
print(hospital_joined.head())
for i, name in enumerate(hospital_joined.columns):
    print("Column #%d : %s" % (i, name))
```

```python
hospital_joined['violent_crime_mod'] = hospital_joined['violent_crime']/(hospital_joined['population_crime']/10000.0)
hospital_joined['property_crime_mod'] = hospital_joined['property_crime']/(hospital_joined['population_crime']/10000.0)
hospital_joined['random'] = np.random.randint(0, 1000, size=len(hospital_joined))
hospital_joined['HospitalNum'] = hospital_joined['state'].map(lambda x: hospital_joined.groupby('state').size()[x])
hospital_joined['HospitalNum_mode'] = (hospital_joined['2010 Census Population'])/(hospital_joined['HospitalNum'])
hospital_joined['Population_rate'] = hospital_joined['2010 Census Population']+1/10000.0

hospital_joined['2010 Census Population'].plot.hist();
pop_mean = hospital_joined['2010 Census Population'].mean()
pop_std = hospital_joined['2010 Census Population'].std()

hospital_joined['Population_determ'] = hospital_joined['2010 Census Population'].apply(lambda x: 2 if ((x-pop_mean)**2)**0.5 > pop_std else 1)

hospital_joined['rating_class'] = hospital_joined['rating'].apply(lambda x: 0 if x == 3 else x)
hospital_joined['rating_class'] = hospital_joined['rating_class'].apply(lambda x: -1 if x < 3 and x > 0 else x)
hospital_joined['rating_class'] = hospital_joined['rating_class'].apply(lambda x: 1 if x > 3 else x)

hospital_joined.info()
#hospital_joined.to_csv('hospital_joined.csv')
```

```python
columns0 = ['random']
columns0_1 = ['mortality_score']
columns1_0 = ['2010 Census Population']
columns1 = ['violent_crime', 'property_crime']
columns2 = ['state_funding_total', 'overall_emp', 'overall_sal', 'investment_total']
columns2_1 = ['pharm_emp', 'pharm_sal', 'meddev_emp', 'meddev_sal', 'res_emp', 'res_sal']
columns3 = ['w_process', 'w_coord', 'w_safety', 'w_cost', 'overall_score']
columns_ex = ['mortality_score', 'safety_score', 'readmission_score', 'experience_score', 'timeliness_score', 'efficient_score']

cols = [columns0, columns0_1, columns1_0, columns1, columns2+columns2_1, columns3, columns_ex]

X = hospital_joined.loc[:, hospital_joined.columns[13:-1]]
y = (hospital_joined['rating_class'])
X0 = hospital_joined.loc[:, columns_ex+columns3]
y0 = hospital_joined['rating_class']



X_train0, X_test0, y_train0, y_test0 = train_test_split(X0, y0, test_size=0.3, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(hospital_joined.groupby('rating').size())
print(hospital_joined.groupby('rating_class').size())
```

```python
modelrdf = GradientBoostingClassifier()
modelrdf.fit(X_train, y_train)

modelrdf_pred = modelrdf.predict(X_test)

print(sns.heatmap(confusion_matrix(y_test, modelrdf_pred, labels=[-1, 0, 1]), annot=True))

feature_imp = modelrdf.feature_importances_
features_comb = zip(hospital_joined.columns[13:-1], feature_imp)

#print(features_comb)
#for i, val in enumerate(sorted(features_comb, key=lambda x: x[1], reverse=True)):
#    print(val[0], val[1])
```

```python
def scoreSummary(name, model, x_tr, x_te, y_tr, y_te):
    model.fit(x_tr, y_tr)
    pred = model.predict(x_te)
    print(name)
    #print(model.feature_importances_)
    print(confusion_matrix(y_te, pred))
    print("Correct Prediction: %d" % (y_te==pred).sum())
    print("Incorrect Prediction: %d" % (y_te!=pred).sum())
    print("Accuracy Score: %f" % accuracy_score(y_te, pred))
    print

def scoreSummary_acc(df, cols, name, model):
    reports = []
    print(name)
    for col in cols:
        belows = []
        aves = []
        aboves = []
        totals = []
        x = df.loc[:, col]
        y = df['rating_class']
        x_tr, x_te, y_tr, y_te = train_test_split(x, y, test_size=0.3, random_state=42)
        model.fit(x_tr, y_tr)
        pred = model.predict(x_te)
        #pred_prob = model.predict_proba(x_te)
        #acc_scores.append((cols.index(col)+1, accuracy_score(y_te, pred)))
        #print(col)
        #print("Log loss: %f" % log_loss(y_te, pred_prob))
        report = classification_report(y_te, pred, target_names=['Below', 'Average', 'Above'])
        print(report)
        report_each = [x for x in report.strip().split(' ') if x]
        belows = [float(x) for x in report_each[5:8]]
        aves = [float(x) for x in report_each[10:13]]
        aboves = [float(x) for x in report_each[15:18]]
        totals = [float(x) for x in report_each[21:24]]

        test01 = pd.DataFrame(columns=['Precision', 'Recall', 'F1'])
        test01.loc['Below'] = belows
        test01.loc['Average'] = aves
        test01.loc['Above'] = aboves
        test01.loc['Total'] = totals
        
        reports.append(test01)
        
    return(reports)

nb_model = GaussianNB()
results_nb = scoreSummary_acc(hospital_joined, cols, 'GaussianNB', nb_model)

knn_model = KNeighborsClassifier()
results_knn = scoreSummary_acc(hospital_joined, cols, 'KNN', knn_model)

rdf_model = RandomForestClassifier(n_estimators=100)
results_rdf = scoreSummary_acc(hospital_joined, cols, 'RandomForest', rdf_model)

gdb_model = GradientBoostingClassifier()
results_gdb = scoreSummary_acc(hospital_joined, cols, 'GradientBoosting', gdb_model)


vote_model = VotingClassifier(estimators=[('knn', knn_model), ('rf', rdf_model), ('gb', gdb_model)], voting='hard')
results_vote = scoreSummary_acc(hospital_joined, cols, 'Voting', vote_model)
```

```python
frame_nb = pd.DataFrame(columns=['Precision', 'Recall', 'F1'])
frame_knn = pd.DataFrame(columns=['Precision', 'Recall', 'F1'])
frame_rdf = pd.DataFrame(columns=['Precision', 'Recall', 'F1'])
frame_gdb = pd.DataFrame(columns=['Precision', 'Recall', 'F1'])
frame_total = pd.DataFrame(columns=['Precision', 'Recall', 'F1'])

columnNames = ['Random', 'Mortality Rate', 'Population', 'by-State Crime', 'by-State Economics', 'Customer Experience', 'Hospital Quality measures']
for x in range(len(results_nb)):
    frame_nb.loc[columnNames[x]] = results_nb[x].loc['Total']

frame_nb.plot.bar(title='GaussianNB');

for x in range(len(results_knn)):
    frame_knn.loc[columnNames[x]] = results_knn[x].loc['Total']
    
frame_knn.plot.bar(title='KNeighbors');

for x in range(len(results_rdf)):
    frame_rdf.loc[columnNames[x]] = results_rdf[x].loc['Total']
    
frame_rdf.plot.bar(title='Random Forests');
    
for x in range(len(results_gdb)):
    frame_gdb.loc[columnNames[x]] = results_gdb[x].loc['Total']
    
frame_gdb.plot.bar(title='Gradient Boosting');
```
